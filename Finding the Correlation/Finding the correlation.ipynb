```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("all_data.csv")
df.head()
```

```python
# Find correlation matrix

matrix = df.corr().round(3)
#low_matrix = np.triu(matrix)
#sns.heatmap(matrix, mask = low_matrix, annot=True, cmap="PiYG")
sns.heatmap(matrix, annot=True, cmap="PiYG")
plt.show()
```

```python
def plot(x, y, title_x, title_y):
    
    # adds the title
    plt.title('Correlation between ' + title_x + " and " + title_y)
    
    # plotting the data
    plt.scatter(x, y)

    # This will fit the best line into the graph
    plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='red')
    
    # Labelling axes
    plt.xlabel(title_x)
    plt.ylabel(title_y)
    
    plt.show()
```

```python
# Date is not important
cols = df.columns.tolist()[1:]

i=0
while i < len(cols):
    j = i+1
    while j < len(cols):
        plot(df[cols[i]], df[cols[j]], cols[i], cols[j])
        j+=1
    i+=1
```
