```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read dataset
df = pd.read_excel("GasConsumption.xlsx")
```

```python
# Remove Unnamed columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Drop first row
df = df.drop(index=df.index[0:2], axis=0)

# Drop rows that are 0
df = df[df['2009'] != 0]

# Change ":" to NaN
df = df.replace(":", np.nan)

# Set index
df.set_index('TIME', inplace=True)
df.index.name = None

# Transpose Dataframe
df = df.T

# Reset index
df.reset_index(inplace = True)

# Change column name
df.rename(columns = {'index':'Year'}, inplace = True)
df.rename(columns = {'Germany (until 1990 former territory of the FRG)':'Germany'}, inplace = True)
df.rename(columns = {'United Kingdom':'UK'}, inplace = True)
df.rename(columns = {'Bosnia and Herzegovina':'Bosnia'}, inplace = True)



display(df)
```

```python
axes = df.plot(x = "Year", xlabel = "", sharex=False, subplots=True, layout=(5,7), figsize=(20,20), color='blue');
axes = axes.ravel()

for ax_id,ax in enumerate(axes):
    if ax_id == 0:
        ax.set_ylabel('Gas Consumption')
        ax.set_xlabel('Year')

plt.show()
```
