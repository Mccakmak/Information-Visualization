```python
import pandas as pd                 # Library used for exporting wikipedia data in Dataframe format and filtering operations.
import matplotlib.pyplot as plt     # Library used for plotting.
```

```python
wiki_link = "https://en.wikipedia.org/wiki/List_of_Nvidia_graphics_processing_units#GeForce_8_(8xxx)_series"

# Gather data from html
all_tables = pd.read_html(wiki_link)

# Tables needed in this assignment
gf8_table_no, gf8_feature_no = 11, 12

# Total tables
print(len(all_tables))
```

```python
# Uncleaned Graphic Card Table
gf8_main = all_tables[gf8_table_no]
gf8_main
```

```python
# Uncleaned Feautre Table
gf8_feat = all_tables[gf8_feature_no]
gf8_feat
```

```python
# Get rid of 2D column for the Graphic Card Table
gf8_main.columns = [col[1] for col in gf8_main.columns] 

# Drop last 2 row because they are column names again.
gf8_main.drop(gf8_main.tail(2).index, inplace = True)
gf8_main
```

```python
# Get rid of 2D column for the Feature table
gf8_feat.columns = [col[1] for col in gf8_feat.columns] 
gf8_feat
```

```python
# Transform Booleans to numbers in Feature Table

gf8_feat = gf8_feat.replace(['Yes', 'No'],[1,0])

# Change the Model names as same in Table 1

gf8_feat = gf8_feat.replace(['GeForce 8800 GS (G92)', 'GeForce 8800 GTS Rev. 2 (G80)', 'GeForce 8800 GT (G92)' ],['GeForce 8800 GS','GeForce 8800 GTS 112 (G80)', 'GeForce 8800 GT'])
gf8_feat
```

```python
# Drop not important columns
gf8_main.drop(gf8_main.columns.difference(['Model','Launch', 'Code name', 'Core (MHz)', 'Shader (MHz)', 'Memory (MHz)', 'Pixel (GP/s)', 'Texture (GT/s)', 'Bus type']), 1, inplace=True)

# Get rows that has GDR3 memory
gf8_main = gf8_main[gf8_main['Bus type'].str.contains('GDDR3')]
gf8_main
```

```python
# Edit weird values with maximum values in Table 1
gf8_main['Memory (MHz)'] = gf8_main['Memory (MHz)'].replace('400700','700') 
gf8_main['Memory (MHz)'] = gf8_main['Memory (MHz)'].replace('700 (256)900 (512, 1024)','900')
gf8_main
```

```python
# Include models same as Table 1 for Feature table

gf8_feat = gf8_feat[gf8_feat['Model'].isin(gf8_main['Model'].values)]
gf8_feat
```

```python
# Plots the graph of models according to data given
def graph_model(df, y_val, y_label, x_label):
    
    # All model names
    model_names = df["Model"].values
    
    # Random color
    #cmap = plt.cm.tab10
    #colors = cmap(np.arange(len(model_names)) % cmap.N)
    
    # Colors for performances
    colors = [{p<-80: '#8c2f23', -80<=p<-60: '#b63827', -60<=p<-40: '#b26f2b', -40<=p<-20: '#db8731', -20<=p<0: '#b9b635', 0<=p<20: '#ddd93e', 20<=p<40: '#c5e24b' , 40<=p<60: '#adc642', 60<=p<80: '#63d041' ,  80<=p: '#479230'}[True] for p in y_val]
    
    plt.bar(model_names, y_val, color=colors)
    # Rotation used for strings on x axis
    plt.xticks(model_names, rotation = 75)
    plt.ylabel(y_label)
    plt.xlabel(x_label)
    plt.show()
```

```python
# Only Clockrate averages

# Means of columns
core_mean = gf8_main["Core (MHz)"].values.astype('int').mean()
shader_mean = gf8_main["Shader (MHz)"].values.astype('int').mean()
memory_mean = gf8_main["Memory (MHz)"].values.astype('int').mean()

clock_rate_mean = (core_mean+shader_mean+memory_mean)/3

# Sums of each row in the data
mhz_sum = gf8_main["Core (MHz)"].values.astype('int') + gf8_main["Shader (MHz)"].values.astype('int') + gf8_main["Memory (MHz)"].values.astype('int')
mhz_sum = mhz_sum/3

# Percentage value that shows the model above or below the average. 
clock_rate_scores = (mhz_sum/clock_rate_mean)*100 - 100

# Avg performance clockrate graph
graph_model(gf8_main, clock_rate_scores, 'Avg. Performance Clock Rate Values %', 'Graphic Card Models')
```

```python
# Only Fillrate averages

# Means of columns
pixel_mean = gf8_main['Pixel (GP/s)'].values.astype('float').mean()
texture_mean = gf8_main['Texture (GT/s)'].values.astype('float').mean()

fillrate_mean = (pixel_mean + texture_mean)/2

# Sums of each row in the data
fill_sum = gf8_main['Pixel (GP/s)'].values.astype('float') + gf8_main['Texture (GT/s)'].values.astype('float')
fill_sum = fill_sum/2

# Percentage value that shows the model above or below the average. 
fill_rate_scores = (fill_sum/fillrate_mean)*100 - 100


# Avg performance fillrate graph

graph_model(gf8_main, fill_rate_scores, 'Avg. Performance Fill Rate Values %', 'Graphic Card Models')
```

```python
# Only Feature averages

# Means of columns
scal_mean = gf8_feat["ScalableLinkInterface(SLI)"].values.astype('int').mean()
sli_mean = gf8_feat["3-WaySLI"].values.astype('int').mean()
pure_mean = gf8_feat["PureVideo HDwith VP1"].values.astype('int').mean()
pure2_mean = gf8_feat["PureVideo 2 with VP2, BSP Engine, and AES128 Engine"].values.astype('int').mean()
pure3_mean = gf8_feat["PureVideo 3 with VP3, BSP Engine, and AES128 Engine"].values.astype('int').mean()
pure4_mean = gf8_feat["PureVideo 4 with VP4"].values.astype('int').mean()
comp_mean = gf8_feat["Computeability"].values.astype('float').mean()


feat_mean = (scal_mean+sli_mean+pure_mean+pure2_mean+pure3_mean+pure4_mean+comp_mean)/7

# Sums of each row in the data
feat_sum = gf8_feat["ScalableLinkInterface(SLI)"].values.astype('int')+gf8_feat["3-WaySLI"].values.astype('int')+gf8_feat["PureVideo HDwith VP1"].values.astype('int')+ gf8_feat["PureVideo 2 with VP2, BSP Engine, and AES128 Engine"].values.astype('int') + gf8_feat["PureVideo 3 with VP3, BSP Engine, and AES128 Engine"].values.astype('int')+gf8_feat["PureVideo 4 with VP4"].values.astype('int')+ gf8_feat["Computeability"].values.astype('float')
feat_sum = feat_sum/7

# Percentage value that shows the model above or below the average.
feat_scores = (feat_sum/feat_mean)*100 - 100

# Avg performance feature graph
graph_model(gf8_feat, feat_scores, 'Avg Performance Feature Values %', 'Graphic Card Models')
```

```python
# Clockrate avg-Fillrate averages

# Percentage value that shows the model above or below the average.
clock_fill_score = (mhz_sum/clock_rate_mean + fill_sum/fillrate_mean)*100 -200

# Avg performance clock-fill rate graph
graph_model(gf8_feat, clock_fill_score, 'Avg Performance Clock-Fill Rate Values %', 'Graphic Card Models')
```

```python
# Clockrate avg-Feature avg

# Percentage value that shows the model above or below the average
clock_feat_score = (mhz_sum/clock_rate_mean + feat_sum/feat_mean)*100 -200

# Avg performance clock-feature rate graph
graph_model(gf8_feat, clock_feat_score, 'Avg Performance Clock-Feature Values %', 'Graphic Card Models')
```

```python
# Fillrate avg-Feature avg

# Percentage value that shows the model above or below the average
fill_feat_score = (fill_sum/fillrate_mean + feat_sum/feat_mean)*100 -200

# Avg performance fillrate-feature graph
graph_model(gf8_feat, fill_feat_score, 'Avg Performance Fillrate-Feature Values %', 'Graphic Card Models')
```

```python
# Clockrate-Fillrate-Feature avg

# Percentage value that shows the model above or below the average
clock_fill_feat_score = (mhz_sum/clock_rate_mean + fill_sum/fillrate_mean + feat_sum/feat_mean)*100 -300

# Avg performance Clockrate-Fillrate-Feature graph
graph_model(gf8_feat, clock_fill_feat_score, 'Avg Performance Clockrate-Fillrate-Feature Values %', 'Graphic Card Models')
```
